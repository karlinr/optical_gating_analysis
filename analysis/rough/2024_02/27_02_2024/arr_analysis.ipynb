## Dose against mean delta phase

```python
doses = [0.1, 0.5, 0.5, 1]
datas = [oog_5dpf_0_1gl, oog_4dpf_0_1gl2, oog_4dpf_0_5gl, oog_4dpf_1_0gl3]

for i, dose in enumerate(doses):
    print(i)
    plt.scatter(doses[i], np.mean(datas[i].delta_phases), color='r')
plt.xlabel('Dose (g/l)')
plt.ylabel('Mean of Delta Phase (radians)')
```

```python
doses = [0.1, 0.5, 0.5, 1]
datas = [oog_5dpf_0_1gl, oog_4dpf_0_1gl2, oog_4dpf_0_5gl, oog_4dpf_1_0gl3]

for i, dose in enumerate(doses):
    print(i)
    plt.scatter(doses[i], np.std(datas[i].delta_phases), color='r')

plt.xlabel('Dose (g/l)')
plt.ylabel('Standard Deviation of Delta Phase (radians)')
```
