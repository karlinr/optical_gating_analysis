```python
import optical_gating_analysis as OG
import matplotlib.pyplot as plt
import numpy as np
import j_py_sad_correlation as jps
import tifffile as tf
```

# Healthy heart

```python
oog_src = r"D:\Data\2012-06-20 13.34.11 vid 2x2 multi phase single plane\brightfield\*tif"
oog = OG.BasicOpticalGating()
oog.sequence_manager.set_source(oog_src)
oog.run()
```

```python
plt.scatter(oog.phases[1::], oog.delta_phases)
plt.xlabel("Phases")
plt.ylabel("Delta phase")
plt.show()
```

```python
sorted = np.argsort(oog.phases)
plt.plot(oog.phases[sorted])
plt.xlabel("Sorted phases")
plt.ylabel("Phase")
plt.show()
```

# Laser-injured heart

```python
oog_src = r"D:\Downloads\Archive\2023-10-30 2hpi_fli1 fish_8_Michellevid\Brightfield - Prosilica\*tif"
oog = OG.BasicOpticalGating()
oog.sequence_manager.set_source(oog_src)
oog.sequence_manager.skip_frames = 0
oog.sequence_manager.set_reference_sequence(r"D:\Downloads\Archive\2023-10-30 2hpi_fli1 fish_8_Michellevid\ref-frames-2023-10-30-17.08.39\*tif")
oog.sequence_manager.reference_period = 36.782587
oog.run()
```

```python
plt.scatter(oog.phases[1::], oog.delta_phases)
plt.xlabel("Phases")
plt.ylabel("Delta phase")
plt.show()
```

```python
sorted = np.argsort(oog.phases)
plt.figure(figsize =(16,9))
plt.scatter(range(oog.phases.shape[0]), oog.phases[sorted], c = np.arange(oog.phases.shape[0])[sorted])
plt.xlabel("Sorted phases")
plt.ylabel("Phase")
plt.show()
```
