```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic(x, a, b, c):
    return a*x**2 + b*x + c

# Generate random sequence of numbers with a linear increase
# Parameters
n = 100
a = 0
b = 0
x = np.linspace(0, 1, n)

# Generate the sequence
y = quadratic(x, a, b, 0) + np.random.normal(0, 1, n)

# Perform moving average
window = 3
y_avg = np.convolve(y, np.array([1, -2, 1]), mode='valid')

# Plot the results
plt.plot(x, y, label='Original')
plt.plot(x[window-1:], y_avg, label='Moving average')
plt.legend()
plt.show()

# Attempt to deconvolve to get original values
y_deconv = np.convolve(y_avg, np.array([1, -2, 1]) / 1, mode='valid')

plt.figure(figsize = (16,9))
plt.plot(x, y, label='Original')
plt.plot(x[window-1:], y_avg, label='Moving average')
plt.plot(x[window*2-2:], y_deconv, label='Deconvolved')
plt.legend()
plt.show()
```
